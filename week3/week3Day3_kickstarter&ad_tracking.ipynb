# Kickstarter project Dataset

```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/makhanov/Dropbox/Yessenov Data Lab/Week 3/Day 3/kickstarter_projects.csv')
```

```python
df.info()
```

```python
# tick is used to deal with space in the name of the column
df.query('`usd pledged` >= 100').head()
```

```python
# 
df.query('state != "suspended"')
```

```python
df['state'].value_counts()
```

```python
df = df.assign(outcome=(df['state']=='successful').astype('int8'))
```

```python
df.head()
```

```python
df['launched'] = pd.to_datetime(df['launched'])
```

```python
df = df.assign(hour=df['launched'].dt.hour,
              day=df['launched'].dt.day,
              month=df['launched'].dt.month,
              year=df['launched'].dt.year)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()

encoded_features = df[['category', 'main_category', 'currency', 'country', ]].apply(encoder.fit_transform)
```

```python
encoded_features
```

```python
df.head()
```

```python
df = df[['goal', 'hour', 'day', 'month', 'year', 'outcome']].join(encoded_features)
df.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('outcome', axis=1),
                                                   df['outcome'],
                                                   test_size=0.3)
rforest = RandomForestClassifier()
rforest.fit(X_train, y_train)
predictions = rforest.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, predictions))
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
rforest.fit(StandardScaler(X_train), y_train)
predictions = rforest.predict(X_test)
```
