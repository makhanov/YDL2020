```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/makhanov/Dropbox/Yessenov Data Lab/Week 3/Day 3/insurance.csv')
```

```python
df.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(['target'], axis=1),
                                                   df['target'],
                                                   test_size=0.3,
                                                   stratify=df['target'])
```

```python
dtree.fit(X_train, y_train)
```

```python
pred = dtree.predict(X_test)
from sklearn.metrics import accuracy_score, classification_report
```

```python
print(accuracy_score(y_test, pred))
```

```python
print(classification_report(y_test, pred))
```

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
from imblearn.under_sampling import TomekLinks
```

```python
tl = TomekLinks(sampling_strategy='majority')
X_tl, y_tl = tl.fit_sample(df.drop(['target'], axis=1),df.target)
```

```python
X = df.drop(['target'], axis=1)
y = df.target
from imblearn.over_sampling import SMOTE
smote = SMOTE(sampling_strategy='minority')
X_smote, y_smote = smote.fit_sample(X, y)
```

```python
from imblearn.combine import SMOTETomek

smotetomek = SMOTETomek()
X_st, y_st = smotetomek.fit_sample(X, y)
```

### RandomOverSampler

```python

X_train, X_test, y_train, y_test = train_test_split(X_over,
                                                   y_over,
                                                   test_size=0.3,
                                                   stratify=y_over
                                                   )
dtree.fit(X_train, y_train)
pred = dtree.predict(X_test)
print(classification_report(y_test, pred))
```

### TomekLinks

```python
X_train, X_test, y_train, y_test = train_test_split(X_tl,
                                                   y_tl,
                                                   test_size=0.3,
                                                   stratify=y_tl
                                                   )
dtree.fit(X_train, y_train)
pred = dtree.predict(X_test)
print(classification_report(y_test, pred))
```

### SMOTE

```python
X_train, X_test, y_train, y_test = train_test_split(X_smote,
                                                   y_smote,
                                                   test_size=0.3,
                                                   stratify=y_smote
                                                   )
dtree.fit(X_train, y_train)
pred = dtree.predict(X_test)
print(classification_report(y_test, pred))
```

### SMOTETomek

```python
X_train, X_test, y_train, y_test = train_test_split(X_st,
                                                   y_st,
                                                   test_size=0.3,
                                                   stratify=y_st
                                                   )
dtree.fit(X_train, y_train)
pred = dtree.predict(X_test)
print(classification_report(y_test, pred))
```

```python
from sklearn.datasets import make_classification
from imblearn.under_sampling import ClusterCentroids
```

```python
X, y = make_classification(n_samples=10000, n_features=3, n_redundant=0, weights=[0.9, 0.1], random_state=42)
```

```python
from imblearn.ensemble import BalancedRandomForestClassifier
bal_random_forest = BalancedRandomForestClassifier()
```

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
```

```python
cv = StratifiedKFold(n_splits=5, random_state=42)
results = cross_val_score(bal_random_forest, X, y, cv=cv)
```

```python
import numpy as np
print(np.mean(results))
```
